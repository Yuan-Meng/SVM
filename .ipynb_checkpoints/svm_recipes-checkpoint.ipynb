# Predict recipe ratings

## Background
The [Epicurious](https://www.kaggle.com/hugodarwood/epirecipes) dataset on Kaggle has "20k recipes listed by recipe rating, nutritional information, and assigned category (sparse)". We can use the given information to predict the average rating of each recipe ranging from 0 to 5.

## Getting data
Let's load the data and take a look.

```python
# Import libraries
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC

%matplotlib inline

import warnings

warnings.filterwarnings("ignore")
```

```python
# Read raw data
raw = pd.read_csv(
    "https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv"
)

# Dimensions
raw.shape

# First 5 rows
raw.head()
```

This dataset has 20,052 rows and 680 columns! It seems we have lots of data cleaning and feature engineering to do.

## Data cleaning

### On the whole set
When cleaning doesn't depend on the specific values of data, we can apply it to the entire dataset. For instance, if some columns have a considerable proportion of missing values, we can drop them from the dataset.

```python
# Count missing values
total = raw.isnull().sum().sort_values(ascending=False)
percent = (raw.isnull().sum() / raw.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=["total", "percent"])
missing = missing.loc[missing["total"] > 0, :]

missing
```

The 4 columns above have over 20% data missing. Let's create a copy of the raw data and drop these columns.

```python
# Copy original dataframe
df = raw.copy()

# Drop columns with too many missing data
df.drop(missing.index.tolist(), axis=1, inplace=True)
```

Since recipe titles are not necessary for making predictions, we can drop the title column as well.

```python
df.drop("title", axis=1, inplace=True)
```

### After train/test split
If cleaning does depend on the exact values of data, we should do it separately for the training data used to fit models and the testing data used to validate models. As a common practice, we can save up 20% of the data for testing later.

#### Train/test split

```python
from sklearn.model_selection import train_test_split

# y is the target variable, recipe ratings
y = df["rating"]

# X is set of features used to predict ratings
X = df.loc[:, df.columns != "rating"]

# Create training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=465
)

df_train = pd.concat([X_train, y_train.reindex(X_train.index)], axis=1)

df_test = pd.concat([X_test, y_test.reindex(X_test.index)], axis=1)
```

#### Median split of ratings
To simplify the prediction problem, especially given the large number of features, we can do a median split of ratings: Ratings higher than or equal to the median are considered "high" (1) and those lower than than the median are considered "low" (0). We can use the training data to establish this cutoff and apply it to the testing data as well.

```python
# Mean of the training data
cutoff = df_train["rating"].median()

# Apply the cutoff to training and testing data
df_train["binary_rating"] = df_train["rating"].apply(lambda x: 1 if x >= cutoff else 0)
df_test["binary_rating"] = df_test["rating"].apply(lambda x: 1 if x >= cutoff else 0)

# Drop off the old column
df_train.drop("rating", axis=1, inplace=True)
df_test.drop("rating", axis=1, inplace=True)

# New target and feature vectors
X_train = df_train.loc[:, df_train.columns != "binary_rating"]
X_test = df_test.loc[:, df_test.columns != "binary_rating"]
y_train = df_train["binary_rating"]
y_test = df_test["binary_rating"]
```

## Data Exploration

### Class imbalance
Since we have hundreds of potential predictors, it's impractical to plot them all and visually detect outliers. Fortunately, since Support Vector Machines (SVM) are robust to outliers, we don't need to worry about this too much. 

However, we need to check for class imbalance in the target variable since SVM may produce [suboptimal results for imbalanced datasets](http://www.cs.ox.ac.uk/people/vasile.palade/papers/Class-Imbalance-SVM.pdf). If the target classes are indeed imbalanced, then as [Chris Albon](https://chrisalbon.com/machine_learning/support_vector_machines/imbalanced_classes_in_svm/) suggested, we need to "increase the penalty for misclassifying minority classes to prevent them from being “overwhelmed” by the majority class". 

```python
sns.countplot(x="binary_rating", data=df_train)
plt.show()
```

It seems high (1) and low (0) ratings are more or less balanced. 

## Feature engineering

### Lasso regularization
As a first pass, we can use Lasso regularization to select features with non-zero coefficients.

```python
from sklearn.linear_model import LassoCV

# Apply Lasso regularization
reg = LassoCV()
reg.fit(X_train, y_train)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" % reg.score(X, y))
coef = pd.Series(reg.coef_, index=X.columns)

# How many features are picked by Lasso?
print(
    "Lasso picked "
    + str(sum(coef != 0))
    + " variables and eliminated the other "
    + str(sum(coef == 0))
    + " variables"
)
```

```python
# Select these features
selected = coef[coef != 0].index.tolist()
X_train = df_train[selected]
X_test = df_test[selected]
```

### Standardize features
To improve performance, we use standardize the features by using their $Z-$scores.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

### Principal Component Analysis (PCA)
To further reduce the feature space, we can perform PCA on the 191 features picked by Lasso. To get a rough estimate of how many components are needed to describe the data, we can plot the cumulative explained variance as a function the number of components selected.

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("number of components")
plt.ylabel("cumulative explained variance");
```

It doesn't seem like the cumulative explained variance hits a point of diminishing return as the number of components increases. For simplicity, however, let's use the first 30 principal components to build SVM models.

```python
n_components = 30

pca = PCA(n_components=n_components)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

## Support Vector Machines

### Train classifier
We can train a binary supporter vector classifier using the 30 principal components we chosen earlier to predict whether a recipe is highly rated or not.

```python
svc = SVC(kernel="linear", C=1.0, random_state=0)

svc.fit(X_train, y_train)
```

### Cross validation
Let's cross validate our classifier on the training dataset. Since it takes too long to run the usual 10-fold cross validation, I opted for 5-fold cross validation and used the `n_jobs=-1` keyword to specify that I want to  recruit all my available cores for this computation (suggested by this [Stack Overflow answer](https://stackoverflow.com/questions/57759943/how-to-speed-up-cross-val-score)).

```python
from sklearn.model_selection import cross_val_score

cross_val_score(svc, X_train, y_train, cv=5, n_jobs=-1)
```

The accuracies are a little under 60% but, still, a huge improvement from about 2% in the tutorial (since login is required, I'm not gonna link it here) where no feature engineering was done before modeling.
